<h3> read data </h3>

```python
df_demand = pd.read_csv(path+'/train.csv')
df_demand.describe().style.background_gradient()
```

```python
df_demand.info()
```

```python
df_demand.isnull().sum()
```

```python
df_demand.head()
```

```python
df_demand.tail()
```

```python
date_sa = df_demand.groupby('date',as_index=False)['sales'].sum()
date_sa
```

<h3> Tiền xử lý và trực quan hóa data </h3>

```python
dfsales = date_sa.reset_index()['sales']
dfsales
```

```python
plt.figure(figsize=(25,13))
plt.plot(dfsales, linewidth=1)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(25,13))
legend = []
for i in range(10):
    store_sales=df_demand.loc[df_demand['store'] == i]
    store_sales=store_sales.groupby('date', as_index=False)['sales'].sum()
    store_sales=store_sales.reset_index()['sales']
    plt.plot(store_sales, linewidth=1)    
    legend.append(('Store '+str(i+1)))
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend(legend, loc='upper left', ncol=1, fancybox=True, shadow=True)
plt.show()
```
